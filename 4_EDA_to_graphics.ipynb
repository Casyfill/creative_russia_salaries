## Datasend Function

```python
def part_datum(part, name=None):
    '''returns nrows, max, min, and 5,25,50,75 and 95 percentiles'''
    part = (part/12).astype(int) # Month
    row = { 'name':name, 'n': len(part), 'max':part.max(), 'min': part.min(), 'std':part.std()}
    for x in (.05, .25, .5, .75, .95):
        row['q%d' % (100*x)] = part.quantile(q=x)
    
    return row

def partition_datim(df, value, pn):
    '''return dataframe of distribution info for partition'''
    
    datum = []
    for name, g in df.groupby(pn):
            datum.append(part_datum(g[value], name))
            
    df =  pd.DataFrame(datum)
    df['ratio'] = df['n'].astype('float')/df['n'].sum()
    df.columns.name = 'N'
    return df
    
```
